```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
import numpy as np

iris = load_iris()
df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
df['species'] = pd.Categorical.from_codes(iris.target, iris.target_names)

df.head()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='sepal length (cm)', y='sepal width (cm)', hue='species', palette='Set1', s=100)
plt.title('Scatter Plot of Sepal Length vs Sepal Width', fontsize=15)
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.legend(title='Species')
plt.show()
```

```python
plt.figure(figsize=(12, 8))
sns.boxplot(x='species', y='sepal length (cm)', data=df)
plt.title('Boxplot of Sepal Length by Species', fontsize=15)
plt.xlabel('Species')
plt.ylabel('Sepal Length (cm)')
plt.show()
```

```python
plt.figure(figsize=(12, 8))
sns.boxplot(x='species', y='sepal width (cm)', data=df)
plt.title('Boxplot of Sepal Width by Species', fontsize=15)
plt.xlabel('Species')
plt.ylabel('Sepal Width (cm)')
plt.show()
```

```python
plt.figure(figsize=(12, 8))
sns.boxplot(x='species', y='petal length (cm)', data=df)
plt.title('Boxplot of Petal Length by Species', fontsize=15)
plt.xlabel('Species')
plt.ylabel('Petal Length (cm)')
plt.show()
```

```python
plt.figure(figsize=(12, 8))
sns.boxplot(x='species', y='petal width (cm)', data=df)
plt.title('Boxplot of Petal Width by Species', fontsize=15)
plt.xlabel('Species')
plt.ylabel('Petal Width (cm)')
plt.show()
```

```python
corr = df.iloc[:, :-1].corr()
plt.figure(figsize=(10, 8))
sns.heatmap(corr, annot=True, cmap='coolwarm', linewidths=0.5, linecolor='black')
plt.title('Correlation Heatmap of Iris Dataset Features', fontsize=16)
plt.show()
```
